```python
import xlsxwriter
import random
import pandas as pd
import openpyxl
from openpyxl.chart import Reference, BarChart
import PyPDF2 
import re
import xml.etree.cElementTree as Et
import docx
```

```python
def task1(workbook_name,worksheet_name,header_list,data):
    
    workbook=xlsxwriter.Workbook(workbook_name)
    worksheet=workbook.add_worksheet(worksheet_name)
    
    worksheet.write(0,0,'Id')
    for idx,header in enumerate(header_list):
        worksheet.write(0,idx+1,str(header).capitalize())

    for idx1,ent in enumerate(data):
        worksheet.write(idx1+1,0,str(idx1+1))
        for idx2,header in enumerate(header_list):
            worksheet.write(idx1+1,idx2+1,ent[header])
    
    workbook.close()

def task2():
    read_file= pd.ExcelFile('shreyas_task1_excel.xlsx')
    print("Excel File have following sheets: ",read_file.sheet_names)
    sheet_to_be_parsed=read_file.sheet_names[0]
    print('Parsing',sheet_to_be_parsed)
    df=read_file.parse(sheet_to_be_parsed)
    print(df)
    return sheet_to_be_parsed
    
def task3():
    wb= openpyxl.load_workbook('shreyas_task1_excel.xlsx')
    ws = wb['s1']
    ws['A5']=4
    ws['B5']='John'
    ws['C5']=65
    wb.save("shreyas_task1_excel.xlsx")
    wb.close()

def task4():
    workbook=xlsxwriter.Workbook('shreyas_task4_chart.xlsx')
    worksheet=workbook.add_worksheet()

    dataseries=[]
    for i in range(0,10,1):
        dataseries.append(random.randint(1,20))
    
    worksheet.write_column("A1",dataseries)

    chart=workbook.add_chart({'type':'line'})

    chart.add_series({'values':'=Sheet1!$A$1:$A$10','name':'My Data'})
    
    worksheet.insert_chart('D1',chart)

    workbook.close()

def task5pdfToexcel():
    pdfFileObj = open('samplepdfforweek1task1.pdf', 'rb') 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
    pageObj = pdfReader.getPage(0) 
    
    str = ""
    for i in range(0,1):
        str += pageObj.extractText()
        
    print(str)
    
    str=str.split('\n')
    str=''.join(str)
    
    print('Parsed Text from Word File: ',str)
    
    y= re.findall("\+\d{2}?[-\s]?\d{10}" , str)
    
    print('Extracted following numbers: ',y)
    
    wb= openpyxl.load_workbook('save_phone_numbers.xlsx')
    ws = wb['pn']
    
    ws['A5']= 'Number extracted from Pdf'
    ws['A6']= y[0]
    
    wb.save('save_phone_numbers.xlsx')
    
    wb.close()
    
    
def task5xmlToexcel():
    tree = Et.parse('sample.xml')
    root=tree.getroot()
    xml_t=''
    for i in root:
        for j in i:
             xml_t+=j.text 
    
    print('Parsed Text from Word File: ',xml_t)
            
    x= re.findall("\+\d{2}?[-\s]?\d{10}" , xml_t)
    
    print('Extracted following numbers: ',x)
    
    wb= openpyxl.load_workbook('save_phone_numbers.xlsx')
    ws = wb['pn']
    ws['E1']='number extracted from xml'
    a=['E2','E3','E4']
    for i in range(len(x)):
        ws[a[i]] = x[i]
            
    wb.save('save_phone_numbers.xlsx')
    wb.close()
    
def task5msToexcel():
    doc = docx.Document("sample_file.docx")
    s=''
    for i in doc.paragraphs:
        s+=i.text
        
    print('Parsed Text from Word File: ',s)
    
    z= re.findall("\+\d{2}?[-\s]?\d{10}" , s)
    
    print('Extracted following numbers: ',z)
    
    wb= openpyxl.load_workbook('save_phone_numbers.xlsx')
    ws = wb['pn']
    
    ws['A8']="number extracted from ms word"
    ws['A9']=z[0]
    
    wb.save('save_phone_numbers.xlsx')
    
    print('Phone numbers saved to excel sheet successfully')
    
    wb.close()
```

```python
decision=input('do you want to start(y(yes)/any key(no))?')
while decision=='y':
    key=input('Enter 1 to perform task1\nEnter 2 to perform task2\nEnter 3 to perform task3\nEnter 4 to perform task4\nEnter 5 to perform task5(Pdf to Excel)\nEnter 6 to perform task5(Xml to Excel)\nEnter 7 to perform task5(Word to Excel)\n')

    if key=='1':
        a=input('Want to create excel file with dummy data?(y)\n Want to enter data for excel file?(n)')
        if a=='y':
            task1('shreyas_task1_excel.xlsx','s1',['name','rollno'],[{'name':'yoshops','rollno':'1'},{'name':'shrey','rollno':'9'},{'name':'shreyas','rollno':'183'}])
        else:
            wbn=input('enter workbook name(without extension)')
            wsn=input('enter worksheet name')
            hl=[]
            s=int(input('enter header_list size'))
            for i in range(s):
                temp=input('enter header value')
                hl.append(temp)
            d=[]
            sz=int(input('enter number of tuples'))
            for i in range(sz):
                temp={}
                for j in hl:
                    temp[j]=input('enter value for '+j+' for '+str(i)+' th sample')
                d.append(temp)
            task1(wbn+'.xlsx',wsn,hl,d)
        print('Excel Sheet created successfully...')
    
    if key=='2':
        print('Excel Sheet parsed successfully: ',task2())
    
    if key=='3':
        task3()
        print("Changes reflects successfully...")
    
    if key=='4':
        task4()
        print('Excel file having 1 chart created successfully...')
    
    if key=='5':
        task5pdfToexcel()
        print("task pdf to excel completed successfully...")
    
    if key=='6':
        task5xmlToexcel()
        print("task xml to excel completed successfully...")

    if key=='7':
        task5msToexcel()
        print("task ms to excel completed successfully...")
    decision=input('do you want to continue(y(yes)/any key(no))?')
print("Work finished")
```
